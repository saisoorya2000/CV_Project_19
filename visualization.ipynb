```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.linalg import norm
```

```python
def visualize(aa,minb=0.9,maxb=2):
    ar = [1, 2, 21, 3, 4, 3, 21, 5, 6, 7, 8, 22, 8, 23, 8, 7, 6, 5, 21, 9, 10, 11, 12, 24, 12, 25, 12, 11, 10, 9, 21, 2, 1, 13, 14, 15, 16, 15, 14, 13, 1, 17, 18, 19, 20];
    xxx = np.zeros(45);
    yyy = np.zeros(45);
    zzz = np.zeros(45);
    xx = np.zeros(45);
    yy = np.zeros(45);
    zz = np.zeros(45);
    for interv in range(30):
        fig = plt.figure()
        ax = fig.gca(projection='3d')
        for j in range(25):
            xx[j] = aa[interv*5,j,0]
            yy[j] = aa[interv*5,j,1]
            zz[j] = aa[interv*5,j,2]
        for j in range(45):
            xxx[j] = xx[ar[j]-1];
            yyy[j] = yy[ar[j]-1];
            zzz[j] = zz[ar[j]-1];
        fig.set_size_inches(8,8)
        minbound = minb
        maxbound = maxb
        ax.plot(xxx,zzz,yyy,linewidth=6,marker='o',mfc='red',ms=5,c='gray')
        ax.auto_scale_xyz([minbound, maxbound], [minbound, maxbound], [minbound, maxbound])
        plt.pause(0.1)
        plt.show()
```

```python
a = np.load('test_ntu_data.npy')
visualize(a[0])
```
